# Preparando os Dados

```python
import pandas as pd

import os

CAMINHO = "C:\\Users\\Nitro\\Desktop\\Deep_Learning\\artistas_brasileiros\\Brazilian_Portugese_Corpus"
```

```python
directories = os.listdir(CAMINHO)
directories = [CAMINHO + "\\" + x for x in directories if not ".txt" in x]

directories
```

```python
# Template de excel
colunas = ["Texto", "Artista"]

dic_excel = {}
```

```python
# Lendo arquivos

dic_textos_artistas = {} # Dicionário que contém como chave o nome do autor e como valor o caminho para seu texto

for directory in directories:
    files = os.listdir(directory)

    dic_textos_artistas[directory.split("\\")[-1]] = [directory + "\\" + file for file in files]

```

```python
dic_conteudo_artistas = {key : [] for key in dic_textos_artistas.keys()}

for key in dic_textos_artistas.keys():
    for caminho_texto in dic_textos_artistas[key]:
        with open(caminho_texto, "r", encoding="latin-1") as arquivo:
            conteudo = arquivo.read()
            dic_conteudo_artistas[key].append(conteudo)


```

# Montando o Excel Template

```python
import pandas as pd

dic_excel["autor"] = [ u + '-' + str(x)  for u in dic_conteudo_artistas.keys() for x in range(len(dic_conteudo_artistas[u]))]

dic_excel["textos"] = [dic_conteudo_artistas[x.split("-")[0]][int(x.split("-")[-1])] for x in dic_excel["autor"]]
#print(dic_excel["textos"][1])

print(len(dic_excel["textos"]))
print(len(dic_excel["autor"]))

print(dic_excel.keys())
```

```python
df = pd.DataFrame.from_dict(dic_excel)

df.to_csv("C:\\Users\\Nitro\\PyCharmMiscProject\\Projetos_DeepLearning\\Classifica_textos_por_artista\\database\\artistas_textos.csv")
```

# Pré-processamento

```python
import pandas as pd

CAMINHO = "C:\\Users\\Nitro\\PyCharmMiscProject\\Projetos_DeepLearning\\Classifica_textos_por_artista\\database\\artistas_textos.csv"

df = pd.read_csv(CAMINHO)
df = df.iloc[:, [1, 2]]
lista_autores_tirar = ["Bernardo Guimaraes","Adolfo Caminha", "Manuel Antonio de Almeida"]

df["autor"] = df["autor"].apply(lambda x: x.split("-")[0])
f = [str(x) + "-" + str(y) for x, y in enumerate(df["autor"])]
f = list(filter(lambda x: x.split("-")[-1] not in lista_autores_tirar, f))
indices = list(map(lambda x: int(x.split("-")[0]), f))
df = df.iloc[indices, :]
print(indices)
df
```

```python
print(df["autor"].value_counts())
```

```python
# Remove duplicadas
autores = list(set(df["autor"]))
autores
```

```python
num_caracteres_por_fragmentos = 500
dic_autores_caracteres = {}
for autor in autores:
    df_autor = df[df["autor"] == autor]
    livros = "\n".join(list(df_autor["textos"]))
    livros_fragmentados = [livros[c: c + 500] if len(livros) - c >= 500 else livros[c: len(livros) - c - 1] for c in range(0, len(livros), num_caracteres_por_fragmentos)]
    dic_autores_caracteres[autor] = livros_fragmentados

print(dic_autores_caracteres.keys())
```

```python
# Criar dataframe com dados balanceados
dic_balanceados = {}
dic_balanceados["autor"] = [u + "-" + str(x) for u in dic_autores_caracteres.keys() for x in range(len(dic_autores_caracteres[u]))]
print(len(dic_balanceados["autor"]))
dic_balanceados["texto"] = [dic_autores_caracteres[u.split("-")[0]][int(u.split("-")[-1])] for u in dic_balanceados["autor"]]
print(len(dic_balanceados["texto"]))
print(dic_balanceados["autor"][76])
print(dic_balanceados['texto'][76])

```

```python
df = pd.DataFrame.from_dict(dic_balanceados)
```

```python
df.to_csv("C:\\Users\\Nitro\\PyCharmMiscProject\\Projetos_DeepLearning\\Classifica_textos_por_artista\\database\\artistas_textos_balanceados.csv")
```

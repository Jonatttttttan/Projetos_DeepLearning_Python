```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import random

import os

from IPython.core.pylabtools import figsize
```

# Carregamento de pastas

```python
CAMINHO = "C:\\Users\\Nitro\\Desktop\\Deep_Learning\\Classificacao_Animais\\Dataset"

pastas = os.listdir(CAMINHO)
pastas
```

```python
pastas = [CAMINHO + "\\" + p for p in pastas]
pastas
```

# Visualização de dados

```python
classes = [nome.split("\\")[-1] for nome in pastas]
#pasta = CAMINHO + "\\" + classes
#print(random.sample(classes, 1))
print(type(random.sample(classes, 1)))

plt.figure(figsize=(10, 10))
for i in range(9):
    classe = random.sample(classes, 1)[0]

    caminho_classe = CAMINHO + "\\" + classe

    imagem = random.sample(os.listdir(caminho_classe), 1)[0]


    img_path = os.path.join(caminho_classe, imagem)

    img = mpimg.imread(img_path)

    plt.subplot(1, 9, i+1)
    plt.imshow(img)
    plt.title(classe)
    plt.axis('off')

plt.tight_layout()
plt.show()

```
